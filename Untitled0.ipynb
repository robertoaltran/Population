```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
A_Sale = pd.read_csv("C:/Users/frpro/OneDrive/Documents/STUDY/3.2.3 158333 Data Processing/RPI/CSV Files/Average Sale Price.csv")
A_Rent = pd.read_csv("C:/Users/frpro/OneDrive/Documents/STUDY/3.2.3 158333 Data Processing/RPI/CSV Files/Average weekly rent.csv")
P_Num = pd.read_csv("C:/Users/frpro/OneDrive/Documents/STUDY/3.2.3 158333 Data Processing/RPI/CSV Files/Number of rental properties.csv")
S_Vol = pd.read_csv("C:/Users/frpro/OneDrive/Documents/STUDY/3.2.3 158333 Data Processing/RPI/CSV Files/Sales volume.csv")

A_Sale = A_Sale.replace('-', 0)
A_Sale['Focus Area'] = A_Sale['Focus Area'] + '_average_sale_price'
A_Sale = A_Sale.rename(columns={'Focus Area': 'Focus_Area'})
A_Sale
```

```python
A_Rent = A_Rent.rename(columns={'Number of Bedrooms': 'Focus Area'})
A_Rent = A_Rent[~A_Rent['Focus Area'].str.contains('Total', case=False)]
A_Rent['Focus Area'] = A_Rent['Focus Area'] + '_average_weekly_rent'
A_Rent = A_Rent.rename(columns={'Focus Area': 'Focus_Area'})
A_Rent
```

```python
P_Num = P_Num.rename(columns={'Number of Bedroom': 'Focus Area'})
P_Num = P_Num[~P_Num['Focus Area'].str.contains('Total', case=False)]
P_Num['Focus Area'] = P_Num['Focus Area'] + '_total_property'
P_Num = P_Num.rename(columns={'Focus Area': 'Focus_Area'})
P_Num
```

```python
S_Vol = S_Vol.replace('-', 0)
S_Vol = S_Vol[~S_Vol['Focus Area'].str.contains('Total', case=False)]
S_Vol['Focus Area'] = S_Vol['Focus Area'] + '_sale_vol'
S_Vol = S_Vol.rename(columns={'Focus Area': 'Focus_Area'})
S_Vol
```

```python
results = pd.concat([A_Sale, A_Rent, P_Num, S_Vol])
results = results.fillna(0)
results.set_index('Focus_Area', inplace=True)
results
```

```python
for col in ['0', '1', '2', '3', '4', '5+']:
    A_Sale[col] = A_Sale[col].str.replace(',', '').str.replace('$', '').astype(float)

A_Sale.set_index('Focus_Area', inplace=True)
A_Sale_transposed = A_Sale.transpose()
plt.figure(figsize=(20, 15))
for focus_area in A_Sale_transposed.columns:
    plt.plot(A_Sale_transposed.index, A_Sale_transposed[focus_area], label=focus_area)
plt.title('Average Sale Prices by Focus Area')
plt.xlabel('Number of Bedroom')
plt.ylabel('Average Sale Price')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 15))

for focus_area in A_Sale_transposed.columns:
    plt.bar(A_Sale_transposed.index, A_Sale_transposed[focus_area], label=focus_area, alpha=0.7)

plt.title('Average Sale Prices by Focus Area')
plt.xlabel('Number of Bedroom')
plt.ylabel('Average Sale Price')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
A_Rent.set_index('Focus_Area', inplace=True)
A_Rent = A_Rent.replace('[\$,]', '', regex=True).astype(float)
A_Rent_transposed = A_Rent.transpose()
plt.figure(figsize=(20, 15))
for focus_area in A_Rent_transposed.columns:
    plt.plot(A_Rent_transposed.index, A_Rent_transposed[focus_area], label=focus_area)
plt.title('Average Weekly Rent by Focus Area')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Weekly Rent')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 15))
for focus_area in A_Rent_transposed.columns:
    plt.bar(A_Rent_transposed.index, A_Rent_transposed[focus_area], label=focus_area, alpha=0.7)
plt.title('Average Weekly Rent by Focus Area')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Weekly Rent')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
